```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Telco-Customer-Churn.csv')  # already loaded in your Colab
df.shape, df.head()
df.info()
df.describe(include='all').T
# percent missing per column
(df.isna().mean()*100).sort_values(ascending=False).head(30)
# churn distribution
df['Churn'].value_counts(normalize=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of a numeric column
df['MonthlyCharges'].hist(bins=30)
plt.title('Distribution of Monthly Charges')
plt.xlabel('MonthlyCharges')
plt.ylabel('Count')
plt.show()

# Boxplot to see outliers by churn
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Monthly Charges by Churn')

# Bar chart for a categorical column
df['Contract'].value_counts().plot(kind='bar')
plt.title('Contract Type Distribution')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.show()
```
